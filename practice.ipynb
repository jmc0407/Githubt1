```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout

# 
#raw_df = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/lstm/005930.KS_3MA_5MA.csv')
#raw_df.head()

from google.colab import drive

drive.mount('/content/gdrive/')#gdrive

#[2] Google Drive  working directory 
import os

working_dir = 'dataset'

colab_default_dir = '/content/gdrive/MyDrive/Colab Notebooks'#

original_dir = os.getcwd()

try:
  os.chdir(colab_default_dir) #Google Drive Colab   

  if not os.path.exists(working_dir):
    os.mkdir(working_dir)

  os.chdir(working_dir) #Working  
  print('currnet dir = ', os.getcwd())# 

except Exception as err:#   
  os.chdir(original_dir)
  print(str(err))

raw_df = pd.read_csv('./005930.KS_3MA_5MA.csv')#  
raw_df.head()

#  
plt.title('SAMSUNG ELECTRONIC STOCK PRICE')
plt.ylabel('price')
plt.xlabel('period')
plt.grid()

plt.plot(raw_df['Adj Close'], label = 'Adj Close')

plt.show()

# (Missing Data ,  )
raw_df.describe()#  
#Missing Data 
raw_df.isnull().sum()

# 0 column 
for col in raw_df.columns:#csv   
  if raw_df[col].min() == 0:#  0
    col_name = col
    print(col_name, type(col_name))# 

raw_df.loc[raw_df['Volume']==0]#

# column 0  
for col in raw_df.columns:
  missing_rows = raw_df.loc[raw_df[col] == 0].shape[0]
  print(col + ': ' + str(missing_rows))

# 0 NaN  , Missing Data 
raw_df['Volume'] = raw_df['Volume'].replace(0, np.nan)

print("----------------------------------------------------")

# column 0  
for col in raw_df.columns:
  missing_rows = raw_df.loc[raw_df[col] == 0].shape[0]
  print(col + ': ' + str(missing_rows))

#missing data 
raw_df.isnull().sum()

raw_df.isnull().any()

raw_df.loc[raw_df['Open'].isna()]

#missing data 
raw_df = raw_df.dropna()
raw_df.isnull().sum()

# (Data    )
from sklearn.preprocessing import MinMaxScaler

# scaler dataframe 
scaler = MinMaxScaler()
scaler_cols = ['Open', 'High', 'Low', 'Close', 'Adj Close', '3MA', '5MA', 'Volume']
scaled_df = scaler.fit_transform(raw_df[scaler_cols])
scaled_df = pd.DataFrame(scaled_df, columns = scaler_cols)
print(scaled_df)

#  3MA, 5MA, Adj Close  feature 
# Adj Close 
#   window_size = 40 
def make_sequene_dataset(feature, label, window_size):
  feature_list = []   # feature list
  label_list = []     # label list

  for i in range(len(feature) - window_size):
    feature_list.append(feature[i:i+window_size])# 
    label_list.append(label[i+window_size])#

  return np.array(feature_list), np.array(label_list)

#feature_df, label_df 
feature_cols = ['3MA', '5MA', 'Adj Close']
label_cols = ['Adj Close']

feature_df = pd.DataFrame(scaled_df, columns = feature_cols)
label_df = pd.DataFrame(scaled_df, columns = label_cols)

#DataFrame => Numpy 
feature_np = feature_df.to_numpy()
label_np = label_df.to_numpy()

print(feature_np.shape, label_np.shape)

#   (make_Sequence_dataset)
window_size = 40
X, Y = make_sequene_dataset(feature_np, label_np, window_size)
print(X.shape, Y.shape)

# ,   
#train, test 
#split = int(len(X)*0.95)
split = -200
x_train = X[0:split]
y_train = Y[0:split]

x_test = X[split:]
y_test = Y[split:]

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)

#   
#model 
model = Sequential()
model.add(LSTM(128, activation = 'tanh', input_shape = x_train[0].shape))#lstm    tanh
model.add(Dense(1, activation = 'linear'))

model.compile(loss = 'mse', optimizer = 'adam', metrics = ['mae'])
model.summary()

# (EarlyStopping )
from tensorflow.keras.callbacks import EarlyStopping

early_stop = EarlyStopping(monitor = 'val_loss', patience = 5)

model.fit(x_train, y_train,
          validation_data = (x_test, y_test),
          epochs = 100, batch_size = 16,
          callbacks = [early_stop])

#   ( MAPE , )
pred = model.predict(x_test)

plt.figure(figsize = (12, 6))
plt.title('3MA + 5MA + Adjs Close, window_size = 40')
plt.ylabel('adj close')
plt.xlabel('period')
plt.plot(y_test, label = 'actual')
plt.plot(pred, label = 'prediction')
plt.grid()
plt.legend(loc = 'best')

plt.show()

# (MAPE)
print(np.sum(abs(y_test-pred) / y_test) / len(x_test))
```
